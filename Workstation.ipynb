```python
words = []
for sub in subs:
    doc = nlp(sub.text)
    words += [w.text.lower() for w in doc if w.pos_ not in ['PUNCT', 'SPACE', 'X']]
print(len(words))
```

```python
## All english subtitles
import glob
words = []
for sub_file in sorted(glob.glob('/vol/work3/maurice/dvd_extracted/*/*.en.srt')):
    #print(sub)
    subs = pysrt.open(sub_file)
    for sub in subs:
        doc = nlp(sub.text)
        words += [w.text.lower() for w in doc if w.pos_ not in ['PUNCT', 'SPACE', 'X']]
words = list(set(words))
print(len(words))
nb_errors = 0
for w in words:
    if w not in nlp.vocab:
        #print('error with word', w)
        nb_errors += 1
print(nb_errors)
```
